Import Essential Libraries & Data

```python
## Import Libraries
import pandas as pd
import numpy as np
```

```python
## Specify Data to Import
filepath = '/content/drive/MyDrive/Colab Notebooks/_Data Files/GC Trans List 2024/P2 2024 NEW Updated.csv'
cols_to_keep = ['action','channel_name','date','username']

## Create variable to save the file name without .csv
file_name = filepath.split('/')[-1].split('.')[0]
```

```python
## Import Data
df =  pd.read_csv(filepath, usecols=cols_to_keep) ## Specify columns so that the data pulls in faster
```

Transform the Data to Be Analyzed

```python
## Add Source Column Using the File Name Variable
df['source'] = file_name
```

```python
## Change the date column to date type
df['date'] = pd.to_datetime(df['date'])

## Sort the date column
df = df.sort_values(by='date')

## See all the dates in the date column
df['date'].unique()
```

```python
## Create conditions list
conditions = [
              df['username'].str.contains('OLO'),
              (df['username'].str.contains('Ziosk') | df['username'].str.contains('ZIOSK'))
]

## Create conditional values list
conditional_values = ['OLO','Ziosk']

## Add new column to df based on conditions
df['channel_check'] = np.select(conditions,conditional_values,default='Misc')

## Concatenate [channel_check] & [channel_name]
df['true_channel'] = df['channel_name'] + '_' + df['channel_check']
```

Group Transformed Data & Validate Accuracy

```python
## Set up columns to group data
groupby_cols = ['true_channel','source']

## Group data by columns and get the record count
df_grouped = df.groupby(groupby_cols).size().reset_index(name='qty')

## Preview grouped data
df_grouped.head(10)
```

Save File in Specified Location

```python
## Save df_grouped to csv
df_grouped.to_csv(f'/content/drive/MyDrive/Colab Notebooks/_Data Files/Output Files/{file_name}_grouped.csv',index=False)
```
